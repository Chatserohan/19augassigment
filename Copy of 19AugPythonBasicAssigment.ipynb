2. Create a program that takes a sentence from the user and counts the number
of vowels (a, e, i, o, u) in the string

```python

vowels='aeiou'
text=input('Enter a text ')
count=0
l=[]
for i in text:
    if i in vowels:
        count+=1
        l.append(i)
print('There are',count,'vowels in',text,'which are =',l)
```
